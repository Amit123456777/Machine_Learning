# OHE VS LE VS OE

___
**<u>One-Hot Encoding (OHE)</u>**

* Definition: Converts each unique category into a separate binary column (0 or 1).
* Representation: Each category becomes a new column in the dataset.

<u>Ohe</u>
| Category | Red | Blue | Green |
|----------|-----|------|-------|
| Red      |  1  |  0   |   0   |
| Blue     |  0  |  1   |   0   |
| Green    |  0  |  0   |   1   |

    
* Use cases: For unordered input features (Nominal data).
* Pros: 
    
    * Avoids implying an ordinal relationship between categories.
    * Works well with ML algorithms that don’t assume order (e.g., Logistic Regression, Decision Trees).

* Cons: Increases dimensionality; creates sparse matrices for high-cardinality features.

___
**<u>Label Encoding</u>**

* Definition: Assigns a unique numeric value to each category.
* Representation: Converts categories into a single column with numeric values.
Example:
| Category | Label |
|----------|-------|
| Red      |   0   |
| Blue     |   1   |
| Green    |   2   |

* Use cases: Mostly for target/output labels (classification problems).
* Pros:
    * Simple and memory-efficient.
    * Easy to implement for algorithms that can handle categorical data directly.
* Cons:
    * Implies an ordinal relationship (e.g., 0 < 1 < 2) which might mislead models for input features.



___
**<u>Ordinal Encoding</u>**

* Definition: Maps categories to numeric values based on their natural order.
* Representation: Categories are assigned numbers in increasing order.
Example:
| Category | Ordinal |
|----------|---------|
| Small    |   0     |
| Medium   |   1     |
| Large    |   2     |

* Use cases: For both input features and target labels when categories have a clear, meaningful order.
* Pros: 
    * Captures the natural relationship between categories.
    * Memory-efficient and easy to use with ordered data.
* Cons:
    * Not suitable for unordered categories.
    * Implies order even if it doesn’t exist, leading to incorrect results for nominal data.

# 13) OUTLIER

An outlier is a data point in a data set that is distant from all other observations.
A data point that lies outside the overall distribution of the dataset.

//A data which is far from/ different from the entire dataset is called outlier.

```python
# explantion:
# hmara data ek given range ke andr dekhne ko milta hai.
# but agr hmara data us range se bahar chla jata hai.
# then we call it outlier.

# aur generally outlier use bola jata hai ki hmara data ek normal distribution ke pattern 
# follow krta hai, but agr uski tail jada long chli jate hai to basically we call
# it outlier.
```

```python
# Explanation of figure. image at (2:23) 👆

# our data follows a normal distribution. but if the data is exceed to
# the 3rd sd, then these data's are called outlier.


# this outlier leads to mistake in the many mathematical formula.
# avg, sd, mean.
# bcoz this outlier creates noise in it.
# and also create distruction in the outcome.

# so basically outlier means -> jb ek data kisi bhi ek given range se bahar chla jata hai.
# then this data is called outlier.

# many ml algorithm starts to give wrong predictions due to outlier.

# means there are many ml algorithm that trains towards the  outlier and starts to give the wrong prediction.
# and it can not give the desired output. It gives the wrong answer.

# ex: [5,6, 7, 5.5, 6.5, 7.5, 9.5, 100]
# 100 is outlier in this scenario
# because this is far from the entire dataset.



# whisker plot ki cap ki value (q3+1.5*iqr and q1-1.5*iqr)
# whisker ko hi caps bol rhe hai.
```

# 14) OUTLIER REMOVAL USING IQR

```python
# //
# WHAT IS AN OUTLIER?
# AGR HMARE PASS KOI EK DATA GIVEN RANGE KE ANDR HOTA HAI.
# AND YDI HMARE KOI DUSRA DATA US RANGE SE BAHAR CHLA JATA HAI.

# IN SIMPLE WE HAVE DATA IN A SPECIFIC RANGE.
# BUT KOI DUSRA DATA USI ME AGR US RANGE SE BHUT HI BEYOND/BAHAR CHL RHE HAI.
# THEN IT IS CALLED OUTLIER.

# 1st we will detect outlier as we have discussed.
# and then we will remove it.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.describe()

# descibe se hme pta chl jata hai ki dataset me outlier hone ke chances hai.
# ya nhi.
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x="CoapplicantIncome",data=dataset)
plt.show()

# there are so many outlier present in this column
```

```python
##### NOW, we will remove outlier.
```

```python
# IQR = Q3-Q1
# min = Q1 - (1.5*IQR)
# max = Q3 + (1.5*IQR)

# Data below min and beyond to max is outlier.
```

```python
# a exam of 100 people
# my rank is 4th
# then percentile is 100-rank = 96


# Q1 = 25 percentile of data
# Q3 = 75 percentile of data
```

```python
# understanding figure 👆

# box,
# whiser -> is line attached side of box,
# cap -> min and max point wali line.
```

```python
dataset.shape
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
# we can find quantiles with descibe function also
```

```python
IQR = q3-q1
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range , max_range
```

```python
new_dataset = dataset[dataset["CoapplicantIncome"]<=max_range]

# //box plot 0 se start ho rha hai to outlier max side hai.
# so hmne min range se outlier ni htaya. only max_range se hta diya ka htaya
```

```python
new_dataset.shape

# 596 18 rows is removed
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_dataset)
```
